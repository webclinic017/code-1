```python
df = pd.read_csv('C:\\Users\\29186\\Desktop\\金股.csv')
```

```python
df1 = df[df['所属日期'] == '2021/1/1']
```

```python
def tran(x):
    if x == '2021/1/1':
        return  '2020/12/31'
    if x == '2021/5/1':
        return '2021/4/30'
    return x

df['所属日期'] = df['所属日期'].apply(tran)
# df[df['所属日期'] =='2021/5/1' ]['所属日期'] = '2021/4/30'
df['date'] = pd.to_datetime(df['所属日期'])
# df[df['所属日期'] =='2021/1/1' ]['所属日期']
```
